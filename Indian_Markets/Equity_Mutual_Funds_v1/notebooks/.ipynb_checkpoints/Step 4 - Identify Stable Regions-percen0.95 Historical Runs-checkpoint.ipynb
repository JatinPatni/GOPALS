```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
portfolio_run_dir = r'../data/simulations/portfolio_simulations_historical'
```

```python
# Configs
k = 3
percentile_ = 0.95

```

```python
dfs = []
for year in tqdm(range(2017, 2025)):
    rundf = pd.read_csv(os.path.join(portfolio_run_dir,'k_'+str(k), 'k_'+str(k)+'_'+str(year)+'.csv'), header=None)
    rundf = rundf.drop_duplicates()
    rundf['sharpe'] = rundf[25]/rundf[26]
    percentile_threshold = rundf['sharpe'].quantile(percentile_)
    rundf = rundf[rundf['sharpe']>=percentile_threshold]
    
    dfs.append(rundf)

finaldf = pd.concat([df.iloc[:,:25] for df in dfs], ignore_index=True)
finaldf_wts = finaldf.groupby(list(finaldf.columns)).size().reset_index(name='wt')
print(finaldf_wts['wt'].unique())
max_wt = max(finaldf_wts['wt'].unique())
trow = finaldf_wts[finaldf_wts['wt']==max_wt]
print(len(trow))
```

```python
filterdfs = []
for year in range(2017, 2025):
    ttdf = pd.read_csv(os.path.join(portfolio_run_dir,'n_'+str(n), 'n_'+str(n)+'_'+str(year)+'.csv'), header=None)
    ttdf = ttdf.drop_duplicates()
    ttdf['sharpe'] = ttdf[25]/ttdf[26]
    filterdf = pd.merge(ttdf, trow, on=list(range(25)), how='inner')
    filterdfs.append(filterdf)
    overall_max_sharpe = round(ttdf['sharpe'].max(),3)
    overall_avg_sharpe = round(ttdf['sharpe'].mean(),3)
    overall_median_sharpe = round(ttdf['sharpe'].quantile(0.5),3)
    max_sharpe = round(filterdf.sharpe.max(),3)
    median_sharpe = round(filterdf.sharpe.quantile(0.5),3)
    avg_sharpe = round(filterdf.sharpe.mean(), 3)
    sharpe_pct_thr = round(ttdf['sharpe'].quantile(percentile_),3)
    print('year -->', str(year), 'overall sharpe max-->', overall_max_sharpe, 'overall avg sharpe -->',overall_avg_sharpe , 'overall median sharpe -->',overall_median_sharpe, 'pct_th -->', percentile_, 'pct value -->',sharpe_pct_thr,
          'pf max_sharpe -->', max_sharpe, 'pf avg sharpe -->', avg_sharpe, 'pf median sharpe -->', median_sharpe )
```

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
df = trow.iloc[:,:25]

df_scaled = df

# Step 2: Apply K-Means clustering
n_clusters = 5  # Number of clusters (can adjust as needed)
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(df_scaled)

# Step 3: Apply t-SNE to reduce dimensions to 2D
tsne = TSNE(n_components=2, random_state=42)
df_tsne = tsne.fit_transform(df_scaled)

# Step 4: Create a DataFrame with the 2D t-SNE results and the cluster labels
df_visualization = pd.DataFrame(df_tsne, columns=['TSNE1', 'TSNE2'])
df_visualization['Cluster'] = cluster_labels

# Step 5: Plot the clusters in 2D with different colors
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_visualization['TSNE1'], df_visualization['TSNE2'], c=df_visualization['Cluster'], cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.title('Portfolio Weights Clusters Visualized with t-SNE')
plt.xlabel('TSNE Dimension 1')
plt.ylabel('TSNE Dimension 2')
plt.show()
```

```python
from scipy.spatial import ConvexHull
```

```python
# Function to plot the efficient frontier and portfolio performance
def plot_risk_return(df, ttdf2, year):
    # Filter data for the specific year
   
    # Calculate the efficient frontier (assuming risk-return values already represent possible portfolios)
    # Efficient frontier can be derived as the outer envelope of return vs risk
    df = df.sort_values(by='annual_vol')
    
    # Combine Risk and Return into one array
    points = np.column_stack((df['annual_vol'], df['annual_ret']))
    
    # Compute the convex hull
    hull = ConvexHull(points)


    # Plot the convex hull (only the boundaries, i.e., the efficient frontier)
    for simplex in hull.simplices:
        plt.plot(points[simplex, 0], points[simplex, 1], 'g-', lw=2)


    # ax.scatter(df['annual_vol'], df['annual_ret'], s=1, alpha=0.1, label="All simulated Portfolios")
    
    # # Plot the efficient frontier
    # ax.plot(df['annual_vol'], df['annual_ret'], color='green', label='Efficient Frontier')

    # Plot portfolio performance (example: pick the portfolio with the maximum return for the plot)
    # max_return_portfolio = df.loc[df['annual_ret'].idxmax()]
    for index, trow in ttdf2.iterrows():
        # ax.scatter(trow['annual_vol'], trow['annual_ret'], color='red', s=100, label=f'Cluster {trow['cluster']}', marker='X')
        plt.scatter(trow['annual_vol'], trow['annual_ret'], s=50, label=f'Cluster {int(trow['cluster'])}', marker='X')

    # Set labels and title
    plt.xlabel('Risk (Volatility)')
    plt.ylabel('Return')
    plt.title(f'Risk-Return Plot for Year {year}')
    plt.legend()
    plt.show()

for year in tqdm(range(2017, 2025)):
    
    # Plot for each year
    
    ttdf = pd.read_csv(os.path.join(portfolio_run_dir,'n_'+str(n), 'n_'+str(n)+'_'+str(year)+'.csv'), header=None)
    ttdf = ttdf.drop_duplicates()

    ttdf.rename(columns={25:'annual_ret', 26:'annual_vol'}, inplace=True)
    ttdf2 = pd.merge(ttdf, pf_wts2, on=list(range(25)), how='inner')

    plot_risk_return(ttdf, ttdf2, year)



```

```python
from matplotlib.lines import Line2D
```

```python
# Show these clusters on a sharpe ratio scale
marker_colors = ['blue', 'green', 'orange', 'purple', 'red']
def plot_sharpe_bar(ttdf, ttdf2, year, ax):
    t_percentile = ttdf['sharpe'].quantile(percentile_)
    ax.bar(year, t_percentile, width=0.7, label=f'Year {year}', color='lightgray')

    
    for index, trow in ttdf2.iterrows():
        # Plot the current portfolio Sharpe ratio as a red X marker
        ax.scatter(year, trow['sharpe'], color=marker_colors[index], s=100, zorder=5, label='Current Portfolio Sharpe', marker='X')

fig, axes = plt.subplots(1, 1, figsize=(18, 6), sharey=True)  # 1 row, 'n_years' columns

for year in tqdm(range(2017, 2025)):
    
    # Plot for each year
    
    ttdf = pd.read_csv(os.path.join(portfolio_run_dir,'n_'+str(n), 'n_'+str(n)+'_'+str(year)+'.csv'), header=None)
    ttdf = ttdf.drop_duplicates()

    ttdf.rename(columns={25:'annual_ret', 26:'annual_vol'}, inplace=True)
    ttdf['sharpe'] = ttdf['annual_ret']/ttdf['annual_vol']
    
    ttdf2 = pd.merge(ttdf, pf_wts2, on=list(range(25)), how='inner')

    plot_sharpe_bar(ttdf, ttdf2, year, axes)

axes.set_xlabel('Years')
axes.set_ylabel('Sharpe 95th percentile')
axes.set_xticklabels(np.arange(2016, 2026))


handles = [Line2D([0], [0], marker='X', color='w', markerfacecolor=mc, markersize=10, label=f'Cluster {cluster}')
           for cluster, mc in zip(np.arange(0,5), marker_colors)]
axes.legend(handles=handles, title='Cluster Sharpe', loc='upper center')


plt.tight_layout()
plt.show()
```
