### Strategy 2.3: Looks only for recent 3 years of historical data and MC Sims
### Here we choose cluster with best cluster silouette scores across mc simulations as well

```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
portfolio_run_dir = r'../data/simulations/portfolio_simulations_monte_carlo'
portfolio_wts_dir = r'../data/backtests/portfolio_weights'
```

```python
# Part 1: Find portfolio weights for 2020
bt_year = 2024
k = 3
percentile_ = 0.95

fnames = os.listdir(portfolio_run_dir)
dfs = []
for year in tqdm(range(bt_year-3, bt_year)):
    t_fnames = [x for x in fnames if str(year) in x]
    for fn in tqdm(t_fnames):
        rundf = pd.read_csv(os.path.join(portfolio_run_dir,fn), header=None)
        rundf = rundf.drop_duplicates()
        rundf['sharpe'] = rundf[25]/rundf[26]
        percentile_threshold = rundf['sharpe'].quantile(percentile_)
        rundf = rundf[rundf['sharpe']>=percentile_threshold]
    
        dfs.append(rundf)

finaldf = pd.concat([df.iloc[:,:25] for df in dfs], ignore_index=True)
finaldf_wts = finaldf.groupby(list(finaldf.columns)).size().reset_index(name='wt')
print(finaldf_wts['wt'].unique())
max_wt = max(finaldf_wts['wt'].unique())
trow = finaldf_wts[finaldf_wts['wt']==max_wt]
print(len(trow))

```

```python
wt_list = finaldf_wts['wt'].unique()
wt_list.sort()
len(wt_list)
```

```python
for i in range(1,len(wt_list)):
    t_wt = wt_list[-1*i]
    
    trow = finaldf_wts[finaldf_wts['wt']==t_wt]
    print(t_wt, len(trow))
    if len(trow)>=100:
        break
```

```python
max_wt = max(finaldf_wts['wt'].unique())
trow = finaldf_wts[finaldf_wts['wt']==t_wt]
print(len(trow))
```

```python
np.random.seed(42)
df = trow.iloc[:,:25]

df_scaled = df

# Step 2: Apply K-Means clustering
n_clusters = 5  # Number of clusters (can adjust as needed)
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(df_scaled)

# Step 3: Apply t-SNE to reduce dimensions to 2D
tsne = TSNE(n_components=2, random_state=42)
df_tsne = tsne.fit_transform(df_scaled)

# Step 4: Create a DataFrame with the 2D t-SNE results and the cluster labels
df_visualization = pd.DataFrame(df_tsne, columns=['TSNE1', 'TSNE2'])
df_visualization['Cluster'] = cluster_labels

# Step 5: Plot the clusters in 2D with different colors
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_visualization['TSNE1'], df_visualization['TSNE2'], c=df_visualization['Cluster'], cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.title('Portfolio Weights Clusters Visualized with t-SNE')
plt.xlabel('TSNE Dimension 1')
plt.ylabel('TSNE Dimension 2')
plt.show()
```

```python
trow['wt'] = cluster_labels
trow.rename(columns={'wt':'cluster'}, inplace=True)
```

```python
trow.groupby('cluster').size()
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples

sample_silhouette_values = silhouette_samples(df_scaled, cluster_labels)

overall_silhouette_score = silhouette_score(df_scaled, cluster_labels)
print(f"Overall Silhouette Score: {overall_silhouette_score}")

data = df_scaled.copy()
# Calculate average silhouette score per cluster
data['Cluster'] = cluster_labels
data['Silhouette Score'] = sample_silhouette_values
cluster_silhouette_scores = data.groupby('Cluster')['Silhouette Score'].mean()
print("Average Silhouette Score per Cluster:")
print(cluster_silhouette_scores)

# Identify the "best" cluster based on the highest silhouette score
best_cluster = cluster_silhouette_scores.idxmax()
print(f"The best cluster based on silhouette score is: {best_cluster}")
```

```python
portfolio_run_dir_hist = r'../data/simulations/portfolio_simulations_historical'
for t_cluster in range(0,5):
    filterdfs = []
    for year in range(bt_year-3, bt_year):
        ttdf = pd.read_csv(os.path.join(portfolio_run_dir_hist,'n_'+str(n), 'n_'+str(n)+'_'+str(year)+'.csv'), header=None)
        ttdf = ttdf.drop_duplicates()
        ttdf['sharpe'] = ttdf[25]/ttdf[26]
        filterdf = pd.merge(ttdf, trow[trow['cluster']==t_cluster], on=list(range(25)), how='inner')
        avg_sharpe = round(filterdf['sharpe'].mean(),3)
        median_sharpe = round(filterdf['sharpe'].median(),3)
        #print('year-->', str(year), 'cluster-->', t_cluster, 'avg_sharpe-->',avg_sharpe, 'median_sharpe-->', median_sharpe)
        filterdfs.append(filterdf)
    tttdf = pd.concat(filterdfs)
    overall_avg_sharpe = round(tttdf['sharpe'].mean(),3)
    overall_median_sharpe = round(tttdf.sharpe.quantile(0.5),3)
    print('year-->', 'all', 'cluster-->', t_cluster, 'avg_sharpe-->',overall_avg_sharpe, 'median_sharpe-->', overall_median_sharpe)
```

```python
best_cluster=0
```

```python
# Choose cluster 0, as max average sharpe here

# Now choose a portfolio weight from this cluster, Choose the medoid, i.e. point in the cluster
# with minimum average distabce to all other points in the cluster. This is like the most representative data point
from sklearn.metrics import pairwise_distances

cluster_data = trow[trow['cluster']==best_cluster].iloc[:,:25].copy()
# Compute distances between all points in the cluster
distances = pairwise_distances(cluster_data)

# Sum distances for each point and find the index of the minimum
medoid_index = np.argmin(distances.sum(axis=1))
medoid_index = int(medoid_index)  # Convert to a standard int

# Now use this as an index for your DataFrame or array
medoid = cluster_data.iloc[medoid_index]  # if using pandas DataFrame
pf_wt = medoid
```
